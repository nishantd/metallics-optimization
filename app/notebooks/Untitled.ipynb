```python
import pandas as pd
import numpy as np
import os
import json
import keras
```

```python
data_path = r'../data/1/previous_heats_with_properties.json'

with open(data_path, 'r') as datafile:
    data = json.load(datafile)
df = pd.DataFrame(data)
df.head(10)
```

```python
df_1 = df.actual_recipe
df_2 = df.chemistry
print(df_1)
```

```python
ls_1 = list(df_1)
ls_sg = list(df_2)
df_rec = pd.DataFrame.from_dict(ls_1)
df_sg = pd.DataFrame.from_dict(ls_sg)
```

```python
print(df_sg)
```

```python
df.drop(columns=['heat_id', 'heat_seq', 'required_weight', 'tap_weight', 'actual_recipe', 'chemistry',], inplace=True)
df.head(10)
```

```python
merged_df = pd.concat([df, df_rec, df_sg], axis=1)
merged_df.replace('ST1', 1, inplace=True)
merged_df.replace('ST2', 2, inplace=True)
merged_df.replace('ST3', 3, inplace=True)
```

```python
merged_df.head(20)
```

```python

np_arr = merged_df.to_numpy()
x = np_arr[::,:-1]
y = np_arr[::,-1]

x_train = x[:-2]
y_train = y[:-2]

x_test = x[-2:]
y_test = y[-2:]
print(x_test)
print(y_test)
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train, x_train).fit()
print_model = model.summary()
print(print_model)
```

```python
predictions = model.predict(x_test)
    
print(predictions)
```
